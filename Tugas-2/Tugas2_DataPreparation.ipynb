```python
#import library

import pandas as pd
import numpy as np
```

```python
#Membaca dan menampilkan 5 baris awal data

df = pd.read_csv('/content/sample_data/data_tugas 2 (1).csv')
df.head()
```

```python
#Cek missing value dan mengubah '?' menjadi 'NaN'

df.replace('?', np.nan, inplace=True)
missing_values = df.isnull().sum()
missing_values
```

```python
#Mengurutkan jumlah missing value dari yang paling banyak ke paling sedikit

missing_values.sort_values(ascending=False)
```

```python
#Mengganti nilai missing value pada variabel kategorik dengan kategori baru yaitu 'unknown'

kategori_cols = ['make', 'bahan-bakar', 'jumlah-pintu', 'body-style']

for col in kategori_cols:
    df[col] = df[col].fillna('unknown')
```

```python
#Mengisi missing value pada variabel numerik dengan rata-rata (mean)

numerik_cols = ['length', 'width', 'height']

for col in numerik_cols:
    mean_value = df[col].mean()
    df[col] = df[col].fillna(mean_value)
    print(f"Mean {col} = {mean_value}")
```

```python
#

min_val = df['wheel-bases'].min()
max_val = df['wheel-bases'].max()

df['wheel-bases-normalized'] = (df['wheel-bases'] - min_val) / (max_val - min_val)
```

```python
# Melakukan normalisasi Z-score pada variabel numerik
# mengubah data agar memiliki rata-rata 0 dan standar deviasi 1
# memudahkan perbandingan antar variabel

for col in numerik_cols:
    mean = df[col].mean()
    std = df[col].std()
    df[col + '_zscore'] = (df[col] - mean) / std
```

```python
df
```

```python
#Karena masih ada missing value 'NaN' pada variabel normalized-losses
#Jadi perlu mengubah tipe data 'NaN' dari string ke numerik
#Mengganti nilai hilang dengan median

df['normalized-losses'] = pd.to_numeric(df['normalized-losses'])
df['normalized-losses'] = df['normalized-losses'].fillna(df['normalized-losses'].median())
```
