#Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

#Membaca Dataset

```python
df = pd.read_csv("https://raw.githubusercontent.com/GeostatsGuy/GeoDataSets/master/1D_Porosity.csv")
```

#Scatter Plot Regresi

#Menentukan variabel

```python
X = df[['Depth']]
y = df['Nporosity']
```

#Membuat Model Regresi / Prediksi Model / Menghitung MAE, MSE, RMSE (evaluasi eror)


*   MAE  = rata-rata error absolut
*   MSE  = rata-rata error kuadrat
*   RMSE = akar dari MSE   
Semakin kecil nilainya, semakin baik model.


```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred)
```

```python
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)


print('MAE:', mae)
print('MSE:',mse)
print('RMSE:',rmse)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Depth', y='Nporosity', data=df, label='Data Asli')
plt.plot(X, y_pred, color='red', label='Garis Regresi')
plt.title('Visualisasi Model Regresi Linier Sederhana')
plt.xlabel('Depth')
plt.ylabel('Nporosity')
plt.legend()
plt.grid(True)
plt.show()
```

#POLYNOMIAL ORDE 2

```python
poly2 = PolynomialFeatures(degree=2)
X_poly2 = poly2.fit_transform(X)

model2 = LinearRegression()
model2.fit(X_poly2, y)

y_pred2 = model2.predict(X_poly2)
```

```python
mae2 = mean_absolute_error(y, y_pred2)
mse2 = mean_squared_error(y, y_pred2)
rmse2 = np.sqrt(mse2)

print("MAE :", mae2)
print("MSE :", mse2)
print("RMSE :", rmse2)
```

```python
plt.scatter(X, y, label="Data Asli")
plt.plot(X, y_pred2, color='green', label="Polynomial Orde 2")
plt.xlabel("Depth")
plt.ylabel("Nporosity")
plt.legend()
plt.show()
```
