```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import time
from google.colab import files
```

```python
data = []

headers = {
    "User-Agent": "Mozilla/5.0"
}

for page in range(1, 6):
    url = f"https://loginmegastore.com/search?query=kulkas&p=true&page={page}"

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "lxml")

    # ambil hanya card produk
    cards = soup.find_all("div", class_="mantine-Card-root")

    for card in cards:
        try:
            # nama produk
            title_tag = card.find("p")

            # harga
            price = None
            for p in card.find_all("p"):
                if "Rp" in p.text:
                    price = p.text.strip()
                    break

            # gambar
            img_tag = card.find("img")

            if title_tag and price and img_tag:
                data.append({
                    "title": title_tag.text.strip(),
                    "price": price,
                    "image": img_tag.get("src")
                })

        except:
            continue

    time.sleep(2)
    print(soup.prettify())
```

```python
print(len(data))
print(data[:5])
```

```python
print("# Hasil scraping:")
for item in data:
    print(item)

print("Total data produk:", len(data))
```

```python
with open("List_Elektronik.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
```
